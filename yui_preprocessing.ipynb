```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('new_dataset/owid-covid-data-most-recent.csv')
df = pd.DataFrame(data)
```

```python
# dropping null values in target as many countries only update records weekly, not daily
df_new_cases_na_dropped = df.dropna(subset = ['new_deaths'])
print(df_new_cases_na_dropped.shape)

```

```python
# filtering countries by null values after dropping null values in target
null_counts = df_new_cases_na_dropped.groupby('location').agg(lambda x: x.isnull().sum(axis=0))

null_counts['total_null'] = null_counts.sum(axis=1)
top_15_df = null_counts[['total_null']].sort_values(by = 'total_null',ascending=True).iloc[:15]
filtered_countries_list = top_15_df.index.to_list()
print(filtered_countries_list)
```
