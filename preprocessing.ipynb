# Preprocessing

```python
# dropping null values in target as many countries only update records weekly, not daily
df = pd.DataFrame(data)
df_new_cases_na_dropped = df.dropna(subset = ['new_deaths'])
print(df_new_cases_na_dropped.shape)
```

```python
# dropping observations before COVID-19 was detected in the country
data = data[~data['total_cases'].isna()]
# filtering countries by null values after dropping null values in target
null_counts = df_new_cases_na_dropped.groupby('location').agg(lambda x: x.isnull().sum(axis=0))

null_counts['total_null'] = null_counts.sum(axis=1)
top_15_df = null_counts[['total_null']].sort_values(by = 'total_null',ascending=True).iloc[:20]
filtered_countries_list = top_15_df.index.to_list()
data = df_new_cases_na_dropped[df_new_cases_na_dropped['location'].isin(filtered_countries_list)]
print(filtered_countries_list)

# dropping redundant columns 
columns_to_drop = ['new_cases_smoothed',
                   'new_cases_smoothed_per_million',
                   'new_deaths_smoothed',
                   'new_deaths_smoothed_per_million',
                   'excess_mortality_cumulative',
                   'excess_mortality_cumulative_absolute',
                   'excess_mortality_cumulative_per_million',
                   'icu_patients',
                   'hosp_patients',
                   'weekly_icu_admissions',
                   'weekly_hosp_admissions',
                   'total_tests',
                   'new_tests',
                   'new_tests_smoothed',
                   'new_tests_smoothed_per_thousand',
                   'tests_per_case', #this is the inverse of positive_rate
                   'new_vaccinations_smoothed',
                   'total_vaccinations',
                   'people_vaccinated',
                   'people_fully_vaccinated',
                   'total_boosters',
                   'new_vaccinations_smoothed_per_million',
                   'new_people_vaccinated_smoothed']
data = data.drop(columns_to_drop,axis=1)
data.shape
```

```python
# find null more than 50% columns
limitPer = len(data) * .5
data_over50 = data.dropna(thresh=limitPer, axis=1)
data_over50.columns
# redundant columns = new_vaccinations_smoothed_per_million and 
# new_people_vaccinated_smoothed_per_hundred
```

```python
#add our column
data_over50['new_deaths'] = data['new_deaths']
```
